# Plotting Every Column in Extracted Feature File

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the path to your CSV file
csv_path = "../data_processed/nodule_features.csv"

# Load the CSV into a DataFrame
df = pd.read_csv(csv_path)

# Select numeric columns
numeric_columns = df.select_dtypes(include=['number']).columns

# Iterate over each numeric column and plot a boxplot and histogram
for col in numeric_columns:
    plt.figure(figsize=(12, 5))
    
    # Boxplot to visualize potential outliers
    plt.subplot(1, 2, 1)
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    
    # Histogram with KDE to visualize the distribution
    plt.subplot(1, 2, 2)
    sns.histplot(df[col], kde=True)
    plt.title(f"Histogram of {col}")
    
    plt.tight_layout()
    plt.show()
```

## Visualizing Extracted Features Separately

```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Path to the CSV file
csv_path = "../data_processed/nodule_features.csv"

# Load the CSV into a DataFrame
df = pd.read_csv(csv_path)

# Convert the "class" column to string if it's not already
df["class"] = df["class"].astype(str)

# Identify numeric feature columns to visualize
# Exclude identification columns if present (e.g., "nodule_number", "patient_id")
exclude_cols = {"nodule_number", "patient_id"}
numeric_features = [col for col in df.select_dtypes(include="number").columns if col not in exclude_cols]

# Set a seaborn style
sns.set(style="whitegrid")

# For each numeric feature, plot a boxplot with color differentiation by class.
for feature in numeric_features:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x="class", y=feature, data=df, palette="Set2")
    plt.title(f"Boxplot of {feature} by Class")
    plt.xlabel("Class")
    plt.ylabel(feature)
    plt.show()
```

### Histogram

```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Path to the CSV file
csv_path = "../data_processed/nodule_features.csv"

# Load the CSV into a DataFrame
df = pd.read_csv(csv_path)

# Convert the "class" column to string if it's not already
df["class"] = df["class"].astype(str)

# Identify numeric feature columns to visualize
# Exclude identification columns if present (e.g., "nodule_number", "patient_id")
exclude_cols = {"nodule_number", "patient_id"}
numeric_features = [col for col in df.select_dtypes(include="number").columns if col not in exclude_cols]

# Set Seaborn style
sns.set(style="whitegrid")

# Define colors for the two classes
class_palette = {"RealCancerous": "blue", "FakeAddedCancer": "red"}

# For each numeric feature, plot a histogram with different colors for each class
for feature in numeric_features:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=df, x=feature, hue="class", palette=class_palette, kde=True, bins=30, alpha=0.6)
    plt.title(f"Histogram of {feature} by Class")
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.legend(title="Class")
    plt.show()
```

## Distance Distribution Curves (KDE Plots)

```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Path to the CSV file
csv_path = "../data_processed/nodule_features.csv"

# Load the CSV into a DataFrame
df = pd.read_csv(csv_path)

# Convert the "class" column to string if it's not already
df["class"] = df["class"].astype(str)

# Identify numeric feature columns to visualize
# Exclude identification columns if present (e.g., "nodule_number", "patient_id")
exclude_cols = {"nodule_number", "patient_id"}
numeric_features = [col for col in df.select_dtypes(include="number").columns if col not in exclude_cols]

# Set Seaborn style
sns.set(style="whitegrid")

# Define colors for the two classes
class_palette = {"RealCancerous": "blue", "FakeAddedCancer": "red"}

# Plot KDE curves for each numeric feature
for feature in numeric_features:
    plt.figure(figsize=(8, 6))
    sns.kdeplot(data=df, x=feature, hue="class", palette=class_palette, fill=True, common_norm=False, alpha=0.5)
    plt.title(f"Distance Distribution Curve (KDE) of {feature} by Class")
    plt.xlabel(feature)
    plt.ylabel("Density")
    plt.legend(title="Class")
    plt.show()
```
